```python
from fractions import Fraction
from decimal import *
```

```python
getcontext().prec = 10000
```

```python
def is_square_root(iterations):
    Num = 1 / (2 + 0.5)
    for i in range(iterations-1):
        deno = 2 + Num
        Num  = Decimal(1/deno)
    
    numo = Fraction(1 + Num).limit_denominator().numerator
    deno = Fraction(1 + Num).limit_denominator().denominator
    
    if len(list(str(numo))) > len(list(str(deno))):
        return True
    return False
```

```python
from time import time
start = time()

count = 0
for i in range(2,1000):
    if is_square_root(i):
        count += 1

print(count)

end = time()
print("=======================================")
print("time taken : " + str(end - start) + " s")
print("=======================================")
```

```python
def square_root(iterations):
    Num = 1 / (2 + 0.5)
    for i in range(iterations-1):
        deno = 2 + Num
        Num  = Decimal(1/deno)
    
    numo = Fraction(1 + Num).limit_denominator().numerator
    deno = Fraction(1 + Num).limit_denominator().denominator
    
    return (numo,deno)
```

```python
for i in range(2,1000):
    R = square_root(i)
    print(R)
    if len(list(str(R[0]))) > len(list(str(R[1]))):
        print(True)
    else:
        print(False)
    print("==========================")
```
