```python
# To calculate the number of multiplications if we use the binary method
def get_n_binary(target):
    exps         = [1, 2]                      # the first exponents is always 1, 2
    to_add_idx   = -1                          # the first one we want to add is the last exponent
    while exps[-1]!=target:
        # adding the biggest exponent possible
        if (exps[-1]+exps[to_add_idx]) <= target:         
            exps.append(exps[-1]+exps[to_add_idx])
        else:
            to_add_idx -= 1 

    return len(exps)-1        # we subtract one because 1 is not counted
```

```python
def get_new_exp(exps, target):
    new_exps = [exps.copy() for _ in range(len(exps))]                 # make a list for each possibility 
    
    for e in range(len(exps)):
        if not (exps[-1]+exps[e]) > target:
            new_exps[e].append(exps[-1]+exps[e])                           # adding the last exp to each exp
 
    return new_exps
```

```python
# To calculate the min number of multiplications possible
def get_n_min(exps, target):
    bin_res = get_n_binary(target)
    
    for i in range(bin_res-2): # -2 as i've already calculated the first two steps
        exps_temp = []
        
        for new_exp in map(lambda x: get_new_exp(x, target), exps):
            # a test if we've get to a sol
            test = list(filter(lambda x: x[-1]==target, new_exp))
            if len(test)>0:
                return len(min(test, key=len))-1

            exps_temp += new_exp
            
        exps = exps_temp

    exps = list(filter(lambda x: x[-1]==target, exps))

    return len(min(exps, key=len))-1
```

```python
def sum_n_min(limit):
    Sum = 0
    exps = [[1,2,3], [1,2,4]]       # exps needed to be list of lists 
    for k in range(3, limit+1):     # m(1)=0, m(2)=1
        print(k)
        Sum += get_n_min(exps, k)
    
    print(f'For 1 ≤ k ≤ 200, find ∑ m(k) = {Sum+1}')
```

```python
%time sum_n_min(200)
```
