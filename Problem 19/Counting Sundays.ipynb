```python
import itertools
```

```python
Months = {"January":[31],"February":[28,29],"March":[31],"April":[30],"May":[31],"June":[30],"July":[31],"August":[31],"September":[30],"October":[31],"November":[30],"December":[31]}
```

```python
Leap_years = [i for i in range(1904,2001,4)] 
```

```python
Current_Sun = 6
Num_D       = 0
for i in range(1901,2001):
    for j in itertools.cycle(Months):
        if j == "February" and i in Leap_years:
            Current_Sun = ((Months[j][1]-Current_Sun)//7)*7 + Current_Sun + 7 - Months[j][1] 
        else:
            Current_Sun = ((Months[j][0]-Current_Sun)//7)*7 + Current_Sun + 7 - Months[j][0]
        if Current_Sun == 1:
            Num_D += 1
            print(j + " in " + str(i))
        
        if j == "December":
            break               

    
print(Num_D)
```
